# Praktikum Preprocessing Data Sederhana
Notebook ini bertujuan untuk mendemonstrasikan tahapan dasar dalam preprocessing data yang mencakup:
1. Data Cleaning (Penanganan Missing Values)
2. Encoding (Mengubah data kategorikal menjadi numerik)
3. Normalisasi (Menyamakan skala data numerik)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

## 1. Membuat Dataset Dummy
Kita akan membuat dataset fiktif berisi data pelamar kerja untuk memprediksi apakah mereka diterima (Label) berdasarkan fitur-fitur seperti Umur, Gaji yang diharapkan, dan Pendidikan.

```python
data = {
    "Umur": [22, 25, np.nan, 28, 30, 24],
    "Gaji_Harapan": [4000000, 6000000, 5500000, 8000000, np.nan, 4500000],
    "Pendidikan": ["SMA", "S1", "S2", "S1", "SMA", "S1"],
    "Status_Diterima": ["Tidak", "Ya", "Ya", "Ya", "Tidak", "Ya"],
}
df = pd.DataFrame(data)
print("=== Dataset Awal ===")
display(df)
```

## 2. Identifikasi Missing Value


Mencari kolom yang memiliki missing value

```python
df.isna().any(axis=0)
```

Mencari row yang memiliki missing value

```python
df.isna().any(axis=1)
```

```python
df[df.isna().any(axis=1)]
```

Mencari row yang tidak memiliki missing value

```python
df[~df.isna().any(axis=1)]
```

## 3. Data Cleaning (Mengatasi Missing Values)
Pada dunia nyata, data sering kali tidak lengkap (`NaN` atau *Not a Number*). 
Kita akan mengisi nilai yang kosong dengan **rata-rata (mean)** dari masing-masing kolom.

```python
# Mengisi nilai NaN menggunakan nilai rata-rata dari kolom tersebut
df_fillna = df.copy()
df_fillna["Umur"] = df_fillna["Umur"].fillna(df_fillna["Umur"].mean())
df_fillna["Gaji_Harapan"] = df_fillna["Gaji_Harapan"].fillna(
    df_fillna["Gaji_Harapan"].mean()
)

print("=== Setelah Data Cleaning ===")
display(df_fillna)
```

Jika data yang tersedia cukup banyak dan missing value yang ada relatif sedikit, Anda dapat mengabaikan row/data yang memiliki missing value.

```python
df_no_nan = df[~df.isna().any(axis=1)]
df_no_nan
```

## 3. Data Encoding (Kategorikal -> Numerik)
Mesin tidak memahami teks ('SMA', 'S1', 'Ya', 'Tidak'). Kita harus mengubahnya menjadi angka.
- **Label Encoding**: Cocok untuk target/label yang memiliki 2 kelas (Ya=1, Tidak=0).
- **One-Hot Encoding**: Cocok untuk fitur nominal/kategorikal agar tidak ada hirarki angka yang salah kaprah.

```python
# 3.1 Label Encoding untuk Target/Label 'Status_Diterima'
le = LabelEncoder()
df_fillna["Status_Diterima"] = le.fit_transform(df_fillna["Status_Diterima"])

# 3.2 One-Hot Encoding untuk Fitur 'Pendidikan'
df_fillna = pd.get_dummies(df_fillna, columns=["Pendidikan"], dtype=int)

print("=== Setelah Encoding ===")
display(df_fillna)
```

## 4. Normalisasi (Scaling)
Kolom 'Umur' memiliki rentang angka puluhan (20-30), sedangkan 'Gaji_Harapan' jutaan (4.000.000 - 8.000.000). 
Model Machine Learning akan bias terhadap nilai yang besar. Oleh karena itu, kita normalisasi menggunakan **MinMaxScaler** agar rentangnya menjadi 0 hingga 1.

```python
scaler = MinMaxScaler()

# Melakukan fit dan transformasi hanya pada kolom numerik
df_fillna[["Umur", "Gaji_Harapan"]] = scaler.fit_transform(
    df_fillna[["Umur", "Gaji_Harapan"]]
)

print("=== Setelah Normalisasi (Data Siap untuk Machine Learning) ===")
display(df_fillna)
```
